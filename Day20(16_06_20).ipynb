```
import pandas as pd
import numpy as np
import sklearn
```

```
data = pd.read_csv("https://raw.githubusercontent.com/Mounika-Kajjam/Datasets/master/weather_data.csv")
data
```

```
data = data.replace({'temperature':-99999,'windspeed':-99999,'event':'0'},np.nan)
data
```

```
#interpolate()
```

```
int_data = data.interpolate()#works only on numerical columns                   it chooses a value inbetwwen max and min values that is to be replaced inplace of Nan
int_data
```

```
#dropna():drops the columns or rows which have missing values
- in general we should not use this
- using this we lose some information
```

```
drop_data = data.dropna()#bydefault how value is any
drop_data
```

```
drop_data = data.dropna(how = 'any')#any means deletee entire row if any missing value is present
drop_data
```

```
#drop the columns having any missing value
drop_data = data.dropna(how = 'any',axis = 1)#by default axis is 0
drop_data
```

```
drop_all_data = data.dropna(how='all')#if entire row is NaN that row is to be droped
drop_all_data
```

##Handling missing values using scikit learn

```
#SimpleImputer
```

```
from sklearn.impute import SimpleImputer##missing values is how are missing value     strategy is with which value it is to be replaced
imputer = SimpleImputer(missing_values = np.nan,strategy= "mean")
```

```
#it is a 2 step transformation  - mean             they are fit and transform
imputer.fit(data.iloc[:,1:3])#indices of column1 and column 2
```

```
data.iloc[:,1:3] = imputer.transform(data.iloc[:,1:3])##to reflect the changes in data           works only for nuemrical
data
```

```
price = pd.read_csv("https://raw.githubusercontent.com/Mounika-Kajjam/Datasets/master/homeprices.csv")
price
```

```
#Convert Categorical Data to a numerical column  1 represents presence of category   0 represents absence of category
dummy_set = pd.get_dummies(price.town) #or price['town']
dummy_set#dummy columns
```

```
price.town
```

```
merged_data = pd.concat([price,dummy_set])  ##always columns should be passed as a list
merged_data
```

```
merged_data = pd.concat([price,dummy_set],axis = 1)
merged_data
```

```
pd.get_dummies(price,columns=['town'])
```
