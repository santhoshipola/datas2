```
#Merging :inner,outer ,left,right
```

```
df1 = pd.DataFrame({"branch":['hyd','blr','vizag','pune'],
                   "total":[90,96,98,92]})
df1 
```

```
df2 = pd.DataFrame({"branch":['blr','hyd','vizag','chennai'],
                   "distance_from_city":[40,60,12,20]})
df2
```

```
pd.merge(df1,df2,on = "branch",how = 'inner')
```

```
pd.merge(df1,df2,on = "branch")#by default  how is inner
```

```
pd.merge(df1,df2,on = "branch",how = 'outer')#like union
```

```
pd.merge(df1,df2,on = "branch",how = 'left')#df1 is left here
```

```
pd.merge(df1,df2,on = "branch",how = 'right',indicator = True)#df2 is right in braces         by default indicator is false
```

```
data = {"name":['a','b','c','d','e','f'],
       "branch":['hyd','vizag','blr','hyd','vizag','blr'],
       "marks":[90,91,92,93,94,95]}
df = pd.DataFrame(data)
df

```

```
g = df.groupby("branch")##divides into groups based on branch 
g
```

```
for branch,branch_df in g:
  print(branch)
  print(branch_df)
```

```
g.get_group('hyd')
```

```
g.get_group('hyd')["marks"].sum()
```

```
g.get_group('hyd')["marks"].mean()
```

```
g.mean()
```
